```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
```

```python
%matplotlib inline
```

```python
data = pd.read_csv("../data/2013_gpv_user-limited.csv")
```

```python
nuserids = data['user_id'].max()
ndays = 365
sub = data[data['days_since'] < ndays]
gpv = np.zeros((nuserids , ndays), dtype='int')
gpv[sub['user_id'].values - 1, sub['days_since'].values] = sub['gpv'].values
out = gpv[gpv.sum(axis=1) > 0]
out -= out.min(axis=0)
out = out.astype('f') / out.max(axis=0)
```

```python
np.save("../data/2013_gpv_user-limited.02.formatted", out)
```
