```python
%%cython --annotate
import numpy as np
cimport numpy as np
cimport cython

from libc.stdlib cimport malloc, free, abs

cdef struct QuadNode:
    # Keep track of the center of mass
    np.float64_t cum_com[2]
    # If this is a leaf, the position of the particle within this leaf 
    np.float64_t cur_pos[2]
    # The number of particles including all 
    # nodes below this one
    np.uint64_t cum_size
    # Number of particles at this node
    np.uint64_t size
    # Index of the particle at this node
    np.uint64_t point_index
    # level = 0 is the root node
    # And each subdivision adds 1 to the level
    np.uint64_t level
    # Left edge of this node, normalized to [0,1]
    np.float64_t le[2] 
    # The center of this node, equal to le + w/2.0
    np.float64_t c[2] 
    # The width of this node -- used to calculate the opening
    # angle. Equal to width = re - le
    np.float64_t w

    # Does this node have children?
    # Default to leaf until we add particles
    np.uint8_t is_leaf
    # Keep pointers to the child nodes
    QuadNode *children[2][2]
    # Keep a pointer to the parent
    QuadNode *parent
    # Keep a pointer to the node
    # to visit next when visting every QuadNode
    # QuadNode *next

@cython.boundscheck(False)
cdef QuadNode* create_root():
    # Create a default root node
    cdef int ax
    root = <QuadNode*> malloc(sizeof(QuadNode))
    root.is_leaf = True
    root.parent = NULL
    root.level = 0
    root.cum_size = 0
    root.size = 0
    root.w = 1.
    for ax in range(2):
        root.le[ax] = 0.
        root.c[ax] = root.le[ax] + root.w / 2.0
        root.cum_com[ax] = 0.
        root.cur_pos[ax] = -1.
    return root

@cython.boundscheck(False)
cdef QuadNode* create_child(QuadNode *parent, np.uint8_t offset[2]):
    # Create a new child node with default parameters
    cdef int ax
    child = <QuadNode *> malloc(sizeof(QuadNode))
    child.is_leaf = True
    child.parent = parent
    child.level = parent.level + 1
    child.size = 0
    child.cum_size = 0
    child.w = parent.w / 2.0
    for ax in range(2):
        child.le[ax] = parent.le[ax] + offset[ax] * parent.w / 2.0
        child.c[ax] = child.le[ax] + child.w / 2.0
        child.cum_com[ax] = 0.
        child.cur_pos[ax] = -1.
    return child

@cython.boundscheck(False)
cdef inline QuadNode* select_child(QuadNode *node, np.float64_t pos[2]):
    # Find which sub-node a position should go into
    # And return the appropriate node
    cdef np.uint8_t offset[2]
    cdef int ax
    for ax in range(2):
        offset[ax] = (pos[ax] - (node.le[ax] + node.w / 2.0)) > 0.
    return node.children[offset[0]][offset[1]]

@cython.boundscheck(False)
cdef subdivide(QuadNode *node):
    # This instantiates 4 nodes for the current node
    cdef int i = 0
    cdef int j = 0
    cdef np.uint8_t offset[2] 
    node.is_leaf = False
    offset[0] = 0
    offset[1] = 0
    for i in range(2):
        offset[0] = i
        for j in range(2):
            offset[1] = j
            node.children[i][j] = create_child(node, offset)

@cython.boundscheck(False)
cdef insert(QuadNode *root, np.float64_t pos[2], int point_index):
    # Introduce a new point into the quadtree
    # by recursively inserting it and subdividng as necessary
    cdef QuadNode *child
    cdef int i
    cdef int ax
    # Evaluate the new center of mass, weighting the previous
    # center of mass against the new point data
    cdef double frac_seen = <double>(root.cum_size - 1) / (<double> root.cum_size)
    cdef double frac_new  = 1.0 / <double> root.cum_size
    for ax in range(2):
        root.cum_com[ax] *= frac_seen
    for ax in range(2):
        root.cum_com[ax] += pos[ax] * frac_new
    # Increment the total number points including this
    # node and below it
    root.cum_size += 1
    # If this node is unoccupied, fill it.
    # Otherwise, we need to insert recursively.
    if (root.size == 0) & root.is_leaf:
        for ax in range(2):
            root.cur_pos[ax] = pos[ax]
        root.point_index = point_index
    else:
        # If necessary, subdivide this node before
        # descending
        if root.is_leaf:
            subdivide(root)
        # We have two points to relocate: the one previously
        # at this node, and the new one we're attempting
        # to insert
        child = select_child(root, root.cur_pos)
        insert(child, root.cur_pos, root.point_index)
        # Remove the point from this node
        for ax in range(2):
            root.cur_pos[ax] = -1
        root.size = 0
        root.point_index = -1
        # Insert the new point
        child = select_child(root, pos)
        insert(child, pos, point_index)

@cython.boundscheck(False)
cdef compute_non_edge_forces(QuadNode* node, 
                             np.float64_t theta,
                             double sum_Q,
                             np.uint64_t point_index,
                             np.ndarray[np.float64_t, ndim=2] pos_reference,
                             np.ndarray[np.float64_t, ndim=2] force):
    # For every position in the reference array, calculate 
    # difference between every populated cell and the position
    # recording along the way how many points that cell represents
    cdef QuadNode* child
    cdef int ax
    cdef int i, child_i, child_j
    cdef int nrows = pos_reference.shape[0]
    cdef np.float64_t dist2, mult, qijZ
    cdef np.float64_t delta[2]
    
    # There are no points below this node
    # Do not bother to calculate naything
    if node.cum_size == 0: 
        return
    # Do not compute self-interactions
    if node.is_leaf and (node.point_index == point_index):
        return

    # For every particle, compute t-SNE force
    # FIX ME: shouldn't loop over every particle here
    dist2 = 0.0
    for ax in range(2):
        # delta[ax] += (node.le[ax] + node.w / 2.0
        #               - pos_reference[point_index][ax]) 
        delta[ax] += node.le[ax] 
        delta[ax] += node.w / 2.0
        delta[ax] -= pos_reference[point_index][ax]
        dist2 += delta[ax]**2.0
    if node.is_leaf or (node.w / np.sqrt(dist2) < theta):
        qijZ = 1.0 / (1.0 + dist2)
        sum_Q += node.cum_size * qijZ
        mult = node.cum_size * qijZ * qijZ
        for ax in range(2):
            force[point_index][ax] += mult * delta[ax]
    else:
        for child_i in range(2):
            for child_j in range(2):
                child = node.children[child_i][child_j]
                compute_non_edge_forces(child, theta, sum_Q, 
                                    point_index,
                                    pos_reference, force)

    
@cython.boundscheck(False)
cdef create_quadtree_cython(np.ndarray[np.float64_t, ndim=2] pos_array):
    cdef QuadNode* root = create_root()
    cdef int nrows = pos_array.shape[0]
    cdef int i, ax
    cdef np.float64_t row[2]
    for i in range(nrows):
        for ax in range(2):
            row[ax] = pos_array[i][ax]
        # print("inserting point %i" % i)
        insert(root, row, i)
    # Consistency check the quadtree
    cdef int count 
    count = count_points(root)
    count_random = count_points_random(root)
    count_root = root.cum_size
    print("  leaves %i points" % count)
    print("  random %i points" % count_random)
    print("    root %i points" % count_root)
    print("expected %i points" % nrows)

@cython.boundscheck(False)
cdef int count_points(QuadNode* root, int count=0):
    # Walk through the whole tree and count the number 
    # of points at the leaf nodes
    cdef QuadNode* child
    cdef int i, j
    for i in range(2):
        for j in range(2):
            child = root.children[i][j]
            if child.is_leaf and child.cur_pos[0] != -1.:
                count += 1
            elif not child.is_leaf:
                count = count_points(child, count=count)
            # else case is we have an empty leaf node
            # which happens when we create a quadtree for
            # one point, and then the other neighboring cells
            # don't get filled in
    return count

# TODO: Make another check for cell boundaries & point data

@cython.boundscheck(False)
cdef int count_points_random(QuadNode* root, int count=0):
    # Walk through the whole tree and count the number 
    # of points at nodes but randomly choose to either
    # descend are use the node's current count (including all
    # sub nodes)
    cdef QuadNode* child
    cdef int i, j, k
    cdef np.ndarray rands = np.random.random(1e6)
    k = 0 
    for i in range(2):
        for j in range(2):
            child = root.children[i][j]
            k += 1
            if rands[k] < 1e-2:
                count += child.size
            else:
                if child.is_leaf and child.cur_pos[0] != -1.:
                    count += 1
                elif not child.is_leaf:
                    count = count_points(child, count=count)
            # else case is we have an empty leaf node
            # which happens when we create a quadtree for
            # one point, and then the other neighboring cells
            # don't get filled in
    return count

def create_quadtree(pos_array):
    create_quadtree_cython(pos_array)
```

```python
times = []
x = np.linspace(1e2, 1e5, 10)
for num in x:
    pos_array = np.random.random((num, 2))
    t0 = time.time()
    create_quadtree(pos_array)
    t1 = time.time()
    times.append(t1-t0)
```
