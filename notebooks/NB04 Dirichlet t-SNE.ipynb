```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
```

```python
%matplotlib inline
```

```python
data = pd.read_csv("../data/gpv_weekly_2013-06-2014-06.csv")
```

```python
user_id_rank, user_id =data['user_id'].factorize()
```

```python
week_rank, week_id = data['days_since'].factorize()
```

```python
nuserids = user_id_rank.max() + 1
nweeks = week_rank.max() + 1
counts = np.zeros((nuserids , nweeks), dtype='int')
counts[user_id_rank, week_rank] = data['total_payment_count'].values
```

```python
counts
```

```python
_ = mpl.plot(counts, lw=0.1, c='k',alpha=0.1)
mpl.ylim(0,2000)
```

### Dirichlet Transformation

```python
counts
```

```python
def transform(sub):
    row = counts[5]
    trans = np.zeros(counts.shape, dtype='f32')
    for j, row in enumerate(sub):
        samples = np.random.dirichlet(row, size=int(1e3))
        psamp = np.percentile(samples, 1.0, axis=0)
        trans[j, :] = psamp
    return trans
```

```python
transf = transform(counts)
```

```python
np.save("../data/counts.npy", transf)
```
